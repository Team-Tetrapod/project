```python
import pandas as pd
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python

def video_frame(video, fps):
    count=0
    fps_frame=[]
    cap_video=cv.VideoCapture(video)
    while(cap_video.isOpened()):
        ret,frame=cap_video.read()
        if int(cap_video.get(1) % fps == 0):
            fps_frame.append(frame)
            count+=1
    return fps_frame
#     if int(cap_video.get(1) % 30 == 0):
#         cv.imwrite("C:/Users/user/Downloads/cvcap/frame%d.jpg" % count, frame)
#         print("saved frame %d" % count)


#     if ret ==False:
#         break
        
# cap_video.release()
```
